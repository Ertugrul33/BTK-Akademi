## Regresyon Nedir?

```python
#Regresyon:

#Çocukların boyları ile babalarının boyları arasında bir ilişki olabilir mi? (Francis Galton)
#Çocukların boyları toplam veri setindeki ortalamaya yakın olma eğilimi gösterebilir.
#Y = a*x + b

#Francis Galton'ın yıllar önce yaptığı çalışmada bu soruyu soruyor. İki veri arasında ilişki kurmaya çalışıyor.
#İki veri arasında ilişki korelasyon olarak da bulunabilir. Fakat regresyon daha fazla sofistike kısmıdır.
#Çünkü regresyonda şu da yapılabilir: Biz bir modeli güzel bir fonksiyona oturtursak örneğin daha çocuk doğmadan
#babanın boyuna bakılarak çocuğun boyu tahmin edilebilir.

#Grafikteki çizgi aslında grafikteki veri setlerinin çizgiye olan uzaklığının minimize edilmiş halidir.
#Ortalama bir çizgidir ve aslında o bizim modelimizdir.
#%100 doğruluk payı yoktur fakat çok yakın olması istenilmektedir.

#"Regres" kelimesi geri çekmek manasındadır. Yani çocukların boyları ortalamaya yakın bir eğilim gösterebilir.
#Örneğin babanın boyu çok uzunsa çocuğun boyu da o kadar olmayacak, "regres" edecek yani geri çekilecek.
#Yani ortalama bir boy olması beklenecek.

#Regresyon denklemindeki "a" değeri aslında ağırlık, "b" değeri ise "bias"tır.
#Bu denklemi oluşturan bizim nöronlarımız ve sinir ağımız, bütün bu işlemleri yapan da TensorFlow'dur.

#Resim için bakınız: TensorFlow-5-Regresyon
```

## Setler (Veri Setleri)

```python
#TensorFlow'da ne yapılacak?

#Verilerimizi ikiye böleceğiz: Öğrenme Verisi(Train Set) ve Test Verisi(Test Set) olarak.
#Bu bölme işlemini "scikit-learn" kütüphanesi ile yapacağız. (makine öğreniminde çok fazla kullanılan bir kütüphanedir)
#Biz birkaç özelliğini kullanacağız. (ortalama alma, medyanına bakma, hatalarına bakma vs.)
#Sonrasında öğrenme verisindeki verilerimizi eğiteceğiz.
#(nöronlar oluşturulacak, derin ağ oluşturulacak, bu ağ içerisinde bir training yani öğrenim süresi geçirilecek)
#Bu öğrenim sonucunda modelimiz tahminler yapmaya hazır olacak.
#İşte biz ayırdığımız test verisini eğittiğimiz modele verip(gerçek sonucu vermeden) gerçek sonucu tahmin etmeye çalışacak.
#Doğruluk oranına göre yani tahmin ile gerçek sonuç arasındaki farkın durumuna göre
#modelin kullanıp kullanılmamasına karar verilecek.

#Örneğin arabanın özellikleri ve fiyatları olan bir verimiz olsun. Bu verimizin %70'lik kısmını modele verdik ve öğrendi.
#%30'luk kısmını da "sadece arabanın özelliklerini" modele vererek fiyatını tahmin etmeye çalıştı.
#Gerçek fiyat ile tahmin arasındaki farkın durumuna göre modeli kullanabiliriz ya da daha fazla veri ile eğitebiliriz.

#Kontollü deney yapabildiğimiz için bu öğrenmeye "Supervised Learning" denir.
#Öğrenmeyi yaptıktan sonra bunu doğruladığımız bir yapı oluşturmamız gerekecek. -> (maliyet fonksiyonu)

#Resim için bakınız: TensorFlow-6-Setler-1-Tensorflow
```

```python
#Maliyet Fonksiyonu (Cost Function veya Error Function):

#Öğrenme süresince bize yardımcı olur, gerçek veriden ne kadar uzaktayız onu ölçer.
#Ve biz bu maliyet fonksiyonu minimize etmeye çalışırız. (gerçek değere yaklaşmak için)
#Bu fonksiyonun değeri ne kadar düşük olursa modelimiz o kadar iyi çalışıyor demektir.

#Gerçek değer ve tahmin değeri bilindiğinden bu işlemi yapabiliriz.

#z = a*g + b
#f(z) = tahminiDeger (nöron'un tahmini)
#Quadratic Maliyet = sum(gercekDeger - tahminiDeger) ** 2 / n
#Cross Entropy Maliyet = (-1/n) * sum(gercekDeger * ln(tahminiDeger) + (1 - gercekDeger * ln(tahminiDeger))

#Resim için bakınız: TensorFlow-6-Setler-2-Maliyet-Fonksiyonu
```

```python
#Gradient Descent (Gradyan Azalması):

#Bir fonksiyonun minimumunu bulmak için kullandığımız optimizasyon fonksiyonudur.
#Maliyet fonksiyonunu minimize etmek için kullanabiliriz.

#Resimdeki grafiğimiz maliyet fonksiyonumuz olsun. Biz en alttaki noktayı bulmayı amaçlıyoruz. 
#Onu da Gradient Descent yapıyor.
#Fonksiyon resimdeki gibi bu kadar basit olmayabilir.

#O en alttaki noktayı bulmak için de eğimi hesaplayıp aşağı doğru iniyoruz.

#Aşağı inerken ne kadar hızlı veya yavaş inebileceğimizi seçebiliriz (step size).
#Bu aslında öğrenme seviyemiz(learning rate) ile aynı anlama gelir.
#Minimuma hızlı ulaşmak için optimizasyonlar kullanabiliriz.

#Learning rate de optimize edilerek bulunan bir değerdir.
#Çok büyük bir değer verildiğinde hızlı bir şekilde atlaya atlaya iner fakat minimum değeri kaçırabilir.
#Çok küçük bir değer verdiğimizde de eninde sonunda bulur fakat çok yavaş ilerler ve zaman kaybı olur.
#Bu yüzden bu değeri optimizasyon yöntemleri kullanarak bulmalıyız.
#Popüler seçenekler: RMSProp, Adam

#Resme göre en iyi minimuma inen optimizasyon algoritması Adam'dır.
#Fakat istisnalar olabilir. Veri setine göre kullanacağımız algoritmayı değiştirebiliriz.

#Resim için bakınız: TensorFlow-6-Setler-3-Gradient-Descent
#Resim için bakınız: TensorFlow-6-Setler-3-Gradient-Descent-2
#Resim için bakınız: TensorFlow-6-Setler-3-Gradient-Descent-3
```

```python
#Tensorflow Sınıflandırma Örneği:

#Amacımız hangilerinin mavi hangilerinin turuncu olduğunu anlayabiliyor mu bunu test etmek.

#Data: Veri setimizin neye benzediğine bakarak ona göre data kısmından şekli seçmemiz gereken yer.
#Features: Girdimizi belirlediğimiz yer.
#Hidden Layer: Gizli katmanlarımızın sayısını belirlediğimiz yer.
#Weight: Ağın üstüne geldiğinde ağırlıkları gösterir. (Derin öğrenme süreci bu ağırlıkları bulmaya yarayacak)
#Ratio of training to test data: Eğitimin test verisine oranı (verinin ne kadarını eğitime vereceğimiz kısım)
#Noise: Sıkıntı çıkartan, alakasız, modeli bozan verileri belirtme yeri. (verimizin durumuna göre belirtebiliriz)
#Batch size: Eğitmek istediğimiz verimizin sayısını belirttiğimiz yer. (fazla veri varsa artırabiliriz)

#OUTPUT kısmında sağ tarafındaki grafik bizim maliyet fonksiyonumuzdur.

#Epoch sayacını sınıflandırma işlemi bittikten sonra durdurmamız gerekir. Yoksa "overfitting"e yol açabilir.
#Overfitting, biz öyle bir denklem kuruyoruz ki gelecekteki verileri tahmin etmesinden çıkıyor.
#Güncel veri setindekilere uydurmaya çalışıyor. Model %100 doğru sonuç verir fakat gelecekte başka bir veri verdiğimizde
#patlayabilir.
#Yani bizim modelimiz bizim veri setimizi veya bizim veri seti ile aynı olan verilerde doğru çalışıp 
#gelecekte başka bir veri verdiğimizde yanlış sonuç verir. O yüzden verimizi çok fazla eğitmemizin anlamı yok.

#Nöronlarımızın sayısını çok fazla artırmak, maliyetli ve zaman alan bir iştir.
#O yüzden nöronlarımızın sayısını da optimize etmemiz gerekiyor.

#Train loss ile test loss arasındaki mesafe az olmalıdır. Burada loss fonksiyonu bizim maliyet fonksiyonumuzdur.

#Nöronların üzerine gelindiğinde hangi nöronun ne iş yaptığı görülür.
#Örneğin bazıları üst tarafta turuncu olanları sınıflar, bazıları sağ alttakileri sınıflar vs.

#Resim için bakınız: TensorFlow-7-TensorflowSiniflandirmaOrnegi
#Resim için bakınız: TensorFlow-7-TensorflowSiniflandirmaOrnegi-2
#Resim için bakınız: TensorFlow-7-TensorflowSiniflandirmaOrnegi-3
#Resim için bakınız: TensorFlow-7-TensorflowSiniflandirmaOrnegi-4
```

## Veriyi Test/Train Olarak İkiye Ayırma

```python
from sklearn.model_selection import train_test_split

#"sklearn.model_selection" kütüphanesinden "train/test ayırma" fonksiyonunu import et demektir.
#Makine öğreniminde çok fazla kullanılan bir kütüphanedir ve bu gibi işlemleri yapmak için kullanılır.
#Bunu kullanmazsak hesaplamaları kendi elimizde yapmamız gerekecekti.
```

```python
#train_test_split

#Dökümantasyona göre önce özellikleri(x->features) ve fiyatı(y->label) diziye çevirmemiz gerekiyor.
```

```python
# y = wx + b
# y -> label

y = dataFrame["Fiyat"].values #Bu şekilde yazarak "Fiyat" verisini listeye çevirebiliriz.

# x -> feature(özellik)

x = dataFrame[["BisikletOzellik1","BisikletOzellik2"]].values
#values denmedeğinde Pandas serisi, values dendiğinde Numpy dizisi olur.

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state = 15)

#Değişkenlerin bu şekilde yazılması önemle tavsiye edilir ve bu sırayla yazılmalıdır.
#test_size: Yüzde kaçını testlere ayıracağımızı yazmalıyız. (varsayılan olarak 0.33'tür)
#random_state: Burada verilen sayı pek önemli değil. Girilmese de olur.
#(Yaptığı iş ise şudur: İki kişi ile çalıştığımızda eğer diğer kişi de aynı değeri yazmışsa 
#veriler aynı algoritma üzerinden seçilecektir. (yani veriler aynı olacaktır)
#(Sonuçları beraber görmek için kullanılabilir fakat çok da önemli değildir.)
```

```python
from sklearn.preprocessing import MinMaxScaler

#Bu sınıfı çağırmamız gerekiyor. (preprocessing: işlemden önce)
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(x_train)
```

```python
#MinMaxScaler(copy=True, feature_range=(0,1)) --> Eski sürümlerde sonuç bu şekilde çıkıyor.
#Bunun anlamı scaler 0-1 arasına çekildi, dönüştürme için hazır demektir.
#Dönüştürme işlemi de "transform" komutu ile yapılır.
```

```python
x_train = scaler.transform(x_train)
x_test = scaler.transform(x_test)

#Bu işlemler tek satırda da yapılabilir. Bir sonraki örnekte o yapılacak.
```

```python
x_train
```

```python
#Artık verilerimiz TensorFlow'da kullanılmak için hazır.
```

```python
import tensorflow as tf
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

#TensorFlow 2 sürümünden sonra Keras kütüphanesini de kullanılabilir hale geldik.
#Sequential: Model bu sınıftan oluşturulur.
#Katmanlarla çalışmak için ise "from tensorflow.keras.layers import Dense" import edilmelidir.
```

```python
model = Sequential()

model.add(Dense(4, activation="relu")) #4 tane nöron ekle ve aktivasyon fonksyonu "relu" olsun demektir.
model.add(Dense(4, activation="relu"))
model.add(Dense(4, activation="relu"))
#3 tane eklediğimiz için de 3 tane gizli katman olacak.

model.add(Dense(1)) #1 tane çıktı nöronu olsun demektir.

model.compile(optimizer = "rmsprop", loss="mse") #MSE: Hata Kareler Ortalaması
```

```python
#İşlemler tamam. Modeli eğitme zamanı geldi.
```

```python
model.fit(x_train, y_train, epochs = 250) #Modeli bu komutlarla eğitiyoruz.

#Verilerimizden 1 kez geçildiğinde 1 Epoch olur.
#Çok fazla verimiz olmadığı için "batch size"ı girmeye gerek yok.
```

```python
loss = model.history.history["loss"]

#Loss değerlerini bize verir. (sözlük olarak tutulduğu için köşeli parantez içerisine "loss"u yazmamız gerekiyor)
```

```python
sbn.lineplot(x=range(len(loss)), y=loss)

#Grafiği çizdirmek için bu komutları kullandık.
#x eksenine loss uzunluğunu(loss'da kaç tane veri olduğunu -> 250), y eksenine loss'ların kendi değerlerini yazdırdık.
```

```python
#Loss değerleri 0'a çok yaklaştığı için model güzel bir halde bitti ve model başarılı olabilir denilebilir.
#(Grafiğin sonradan artmaması güzel bir haber. Yani gerçek değere çok yaklaştı denilebilir.)
```

```python
trainLoss = model.evaluate(x_train, y_train) #Modeli değerlendir demektir.
```
