```python
import numpy as np
import pandas as pd
```

```python
sozlukVerisi = {"Istanbul" : [30,29,np.nan], "Ankara" : [20,np.nan,25], "Izmir" : [40,39,38]}
havaDurumuDataFrame = pd.DataFrame(sozlukVerisi)

#Bazı veriler boş olduğunda(Missing Data) yapılması gerekenler anlatılıyor. (np.nan boş demek.)
```

```python
havaDurumuDataFrame
```

### 1- Olmayan verileri komple silebiliriz.

```python
havaDurumuDataFrame.dropna()

#dropna metodu, varsayılan olarak bir satırda NaN değeri varsa satırı komple siler.
```

```python
havaDurumuDataFrame

#Bilindiği üzere Data Framemimiz hala değişmedi.
```

```python
havaDurumuDataFrame.dropna(axis = 1)

#Sütun olarak NaN olanları sil dersek dropna içerisine axis = 1 yazmamız gerekiyor.
```

```python
#Fakat 1 tane NaN var diye satırı veya sütunu komple silmek mantıklı değil.
#Fazlaca NaN varsa silmek mantıklı olur.
```

```python
yeniVeri = {"Istanbul" : [30,29,np.nan], "Ankara" : [20,np.nan,25], "Izmir" : [40,39,38], "Antalya" : [45,np.nan,np.nan]}
yeniDataFrame = pd.DataFrame(yeniVeri)
```

```python
yeniDataFrame
```

```python
yeniDataFrame.dropna(axis = 1, thresh = 2)

#thresh argümanı 2 ve üzeri olan NaN'ları kaldır demek. (sütundan kaldırmak istediğimiz için axis = 1 yazdık.)
#Bu veri için satırlarda 2 NaN olanları kaldırsaydık 1 ve 2'de 2 tane NaN olduğu için tek 0 satırı kalacaktı.
```

### 2- Olmayan verilerimize ortalama değer atayabiliriz.

```python
yeniDataFrame
```

```python
yeniDataFrame.fillna(20)

#fillna metodu boş olan verileri içerisinde yazılan değer ile doldur demek.
```

## groupby

```python
#Satırları gruplamaya yarar.
#MultiIndex yapısı gibidir fakat herhangi bir multi index olmadan da kolondan(sütundan) gruplama yapılabilir.
```

```python
maasSozlugu = {"Departman" : ["Yazılım","Yazılım","Pazarlama","Pazarlama","Hukuk","Hukuk"],
              "Calisan İsmi" : ["Ahmet","Mehmet","Atıl","Burak","Zeynep","Fatma"],
                "Maas" : [100,150,200,300,400,500]
              }
```

```python
maasDataFrame = pd.DataFrame(maasSozlugu)
```

```python
maasDataFrame
```

```python
#Departmana göre gruplama yapmaya çalışalım.
```

```python
maasDataFrame.groupby("Departman")
```

```python
#RAM'e kaydedildi dedi.
#Departmana göre gruplama yaptık fakat biz bununla ilgili işlemler yapabilmemiz için bir değişkene eşitlememiz gerekiyor.
```

```python
grupObjesi = maasDataFrame.groupby("Departman")
```

```python
grupObjesi.count()
```

```python
grupObjesi.mean()
```

```python
grupObjesi.min()
```

```python
grupObjesi.max()
```

```python
grupObjesi.describe()

#Verilerimiz ile ilgili genel sonuçları verir. (ortalama, sapma, medyan vs.) (Güzel Metod !!)
```
